```python
import pandas as pd

# Create a dictionary to store the data
data = {
    "Name": ["Kiran", "Arun", "Vijay", "Varun"],
    "Age": [25, 30, 35, 40],
    "Height": [145, 151, 165, 173],
    "Weight": [45, 55, 65, 75]
}

# Create a Pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Age vs Height
plt.scatter(df["Age"], df["Height"])
plt.title("Scatter Plot of Age vs Height")
plt.xlabel("Age (years)")
plt.ylabel("Height (cm)")
plt.show()

# Age vs Weight
plt.scatter(df["Age"], df["Weight"])
plt.title("Scatter Plot of Age vs Weight")
plt.xlabel("Age (years)")
plt.ylabel("Weight (kg)")
plt.show()

# Height vs Weight
plt.scatter(df["Height"], df["Weight"])
plt.title("Scatter Plot of Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()

```

```python
# Distribution of Age
df["Age"].value_counts().plot(kind="bar")
plt.title("Bar Chart of Age Distribution")
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.show()

# Distribution of Height
df["Height"].value_counts().plot(kind="bar")
plt.title("Bar Chart of Height Distribution")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()

# Distribution of Weight
df["Weight"].value_counts().plot(kind="bar")
plt.title("Bar Chart of Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram of Age
plt.hist(df["Age"])
plt.title("Histogram of Age")
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.show()

# Histogram of Height
plt.hist(df["Height"])
plt.title("Histogram of Height")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()

# Histogram of Weight
plt.hist(df["Weight"])
plt.title("Histogram of Weight")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.show()
```
